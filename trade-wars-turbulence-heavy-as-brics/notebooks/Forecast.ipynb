```python
import pandas as pd
from prophet import Prophet

# Filter data for South Africa
sa_currency_df = df[df['Country'] == 'South Africa'][['Year', 'Currency_Value_USD']].copy()

# Prophet needs 'ds' (datetime) and 'y' (value to forecast)
sa_currency_df['ds'] = pd.to_datetime(sa_currency_df['Year'], format='%Y')
sa_currency_df['y'] = sa_currency_df['Currency_Value_USD']

# Keep only required columns
currency_data = sa_currency_df[['ds', 'y']]
```

```python
# Create the model
currency_model = Prophet()

# Fit the model on the currency data
currency_model.fit(currency_data)

# Forecast for the next 10 years
future_currency = currency_model.make_future_dataframe(periods=10, freq='YE')
currency_forecast = currency_model.predict(future_currency)
```

```python
import matplotlib.pyplot as plt

fig = currency_model.plot(currency_forecast)
plt.title("Forecasted Currency Value ZAR-USD  South Africa (20102035)")
plt.xlabel("Year")
plt.ylabel("Currency Value in USD")
plt.grid(True)
plt.tight_layout()
plt.savefig("Forecasted Currency Value ZAR-USD  South Africa (20102035).png")
plt.show()
```
